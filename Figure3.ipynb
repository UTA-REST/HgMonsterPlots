```python
import pylab
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
dat=pd.read_excel("Data/Fig3.xlsx")
dat
```

```python
x1 = list(dat['6A'])
x2 =list(dat['6B'])
x3 = list(dat['6C'])
y1 = list(dat['Unnamed: 1'])
y2 = list(dat['Unnamed: 3'])
y3 = list(dat['Unnamed: 5'])
```

```python
plt.figure(figsize=(5,5),dpi=200)
plt.plot(x1,y1,color='black')
plt.plot(x2,y2,color='DarkRed')
plt.plot(x3,y3,color="DarkBlue")

pylab.ylim(0,0.8)
pylab.xlim(250,750)
pylab.xlabel("Wavelength (nm)")
pylab.ylabel("Fluorescene Intensity (arb)")
pylab.legend(['1A','1B','1C'],loc='upper right')
ax = plt.gcf().gca()
plt.tight_layout()
plt.savefig('Figs/Figure3.png',bbox_inches='tight')
plt.show()
```
