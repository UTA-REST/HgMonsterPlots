```python
import pyequion
sol = pyequion.solve_solution({'MgCl2': 1})
concs=sol.concentrations
```

```python
def def_value(): 
    return []
```

```python
dicts=defaultdict(def_value)

Totals=np.logspace(-3,3,100)
for Total in Totals:
    sol = pyequion.solve_solution({'CuCl2': Total})
    concs=sol.concentrations
    for i in concs.keys():
        dicts[i].append(concs[i])
```

```python
plt.figure(figsize=(5,5),dpi=150)
for d in dicts.keys():
    plt.plot(Totals,dicts[d],label=d)
    plt.loglog()
plt.legend(loc='lower right')
plt.xlabel("Total Mg concentration (mM)")
plt.ylabel("Speciation")
```
