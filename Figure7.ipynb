```python
import pylab
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
dat=pd.read_excel("Data/Metaltestingisolated.xlsx")
dat
```

```python
x = list(dat['Wavelength (nm)'])

y={}

y1 = list(dat['Hg Fresh'])
y2 = list(dat['Ag'])
y3 = list(dat['K'])
y4 = list(dat['Ba'])
y5 = list(dat['Pb'])
y6 = list(dat['Zn'])
y7= list(dat['Ca'])
y8= list(dat['Na'])
y9= list(dat['Mg'])
y10= list(dat['Cd'])
y11= list(dat['Li'])
y12= list(dat['Cs'])
y13= list(dat['Rb'])
y14= list(dat['Me'])
y15= list(dat['Crown'])
y17= list(dat['Cu Fresh'])

Order=['Hg Fresh','Ag','K','Ba','Pb','Zn','Ca','Na','Mg','Cd','Li','Cs','Rb','Me','Crown','Cu Fresh']
Colors={}
for i in range(0,len(Order)):
    y[i]=list(dat[Order[i]])


ColorMap={'Hg Fresh':'#ff0000', 'K':'#ff8000','Cs':'#FFC600','Ag':'#C4D300','Rb':'#00ff00','Ca':'#44D7A8','Na':'#00ffff',
          'Cd':'#0080ff', 'Mg':'#0000ff', 'Li':'#8000ff', 'Zn':'#FC8EAC', 'Me':'#ff0080','Pb':'DarkGreen','Cu Fresh':'black','Ba':'dimgrey','Crown':'dimgrey'}

ChargeStates={'Hg Fresh':'$^{2+}$', 'K':'$^{+}$','Cs':'$^{+}$','Ag':'$^{+}$','Rb':'$^{+}$','Ca':'$^{2+}$','Na':'$^{+}$',
          'Cd':'$^{2+}$', 'Mg':'$^{2+}$', 'Li':'$^{+}$', 'Zn':'$^{2+}$', 'Me':'$^{+}$','Pb':'$^{2+}$','Cu Fresh':'$^{2+}$','Ba':'$^{2+}$','Crown':""}
Names={}
a=[]
for i in range(0,len(Order)):
    Names[i]=Order[i]+ChargeStates[Order[i]]



    if(Order[i]=='Me'):
        Names[i]=r'NMe$_4^+$'
    if(Order[i]=='Hg Fresh'):
        Names[i]=r'Hg$^{2+}$'
    if(Order[i]=='Cu Fresh'):
        Names[i]=r'Cu$^{2+}'
    a.append( mpatches.Patch(facecolor=ColorMap[Order[i]], label = Names[i], linewidth = 0.5, edgecolor = 'black'))


```

```python
plt.figure(figsize=(5,5),dpi=200)
for i in range(0,len(Order)):
    plt.plot(x,y[i],color=ColorMap[Order[i]],label=Names[i])

pylab.ylim(0,0.135)
pylab.xlim(350,550)
pylab.xlabel("Wavelength (nm)")
pylab.ylabel("Fluorescene Intensity (arb)")
pylab.legend(handles=a,loc='upper right',ncol=4,fontsize=9.5)
ax = plt.gcf().gca()
ax.annotate('Sensor + $\mathregular{Hg^{2+}}$', 
            xy=(420, 0.07),
            xycoords='data',
            xytext=(370, 0.09), 
            va='center',
            ha='left',
            arrowprops={'arrowstyle': '-|>', 'lw': 2})
ax.annotate('Sensor + \nOther Metals', 
            xy=(490, 0.055),
            xycoords='data',
            xytext=(540, 0.08), 
            va='center',
            ha='right',
            arrowprops={'arrowstyle': '-|>', 'lw': 2})
plt.tight_layout()
plt.savefig('Figs/Figure7.png',bbox_inches='tight')
plt.show()
```
