```python
import pylab
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
dat=pd.read_excel("Data/Solvents_Isolated.xlsx")
dat
```

```python
x = list(dat['Wavelength (nm)'])
x2 = list(dat['Wavelength EtOH'])
y1 = list(dat['DMSO'])
y2 = list(dat['MeOH'])
y3 = list(dat['DCM'])
y4 = list(dat['EtOH'])
y5 = list(dat['THF'])
```

```python
a1 = mpatches.Patch(facecolor='red', label='DMSO', linewidth = 0.5, edgecolor = 'black')
a2 = mpatches.Patch(facecolor='blue', label = 'MeOH', linewidth = 0.5, edgecolor = 'black')
a3 = mpatches.Patch(facecolor='green', label = 'DCM', linewidth = 0.5, edgecolor = 'black')
a4 = mpatches.Patch(facecolor='purple', label = 'EtOH', linewidth = 0.5, edgecolor = 'black')
a5 = mpatches.Patch(facecolor='orange', label = 'THF', linewidth = 0.5, edgecolor = 'black')
```

```python
plt.figure(figsize=(5,5),dpi=200)
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='blue')
plt.plot(x,y3,color='green')
plt.plot(x2,y4,color='purple')
plt.plot(x,y5,color='orange')
pylab.ylim(0,0.14)
pylab.xlim(350,600)
pylab.xlabel("Wavelength (nm)")
pylab.ylabel("Fluorescene Intensity (arb)")
pylab.legend(handles=[a1,a2,a3,a4,a5],loc='upper right') 
plt.tight_layout()
plt.savefig('Figs/Figure6.png',bbox_inches='tight')
plt.show()
```
